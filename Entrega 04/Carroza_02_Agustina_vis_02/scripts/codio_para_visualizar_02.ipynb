```python
import pandas as pd

# Especifica la ruta al archivo en Google Drive
file_path = '/content/drive/MyDrive/Netflix subscription fee Dec-2021 - Netflix subscription fee Dec-2021.csv.csv'

# Carga el archivo CSV
df = pd.read_csv(file_path)

# Muestra las primeras filas del DataFrame
df.head()
```

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
for pais in df['Country'].unique():
    data = df[df['Country'] == pais]
    plt.scatter(data['Total Library Size'], data['Cost Per Month - Basic ($)'], label=pais)

plt.xlabel('Total Library Size')
plt.ylabel('Cost Per Month - Basic ($)')
plt.title('Relación entre tamaño de la biblioteca y costo por mes por país en Netflix')
plt.legend()
plt.grid(True)
plt.show()
```

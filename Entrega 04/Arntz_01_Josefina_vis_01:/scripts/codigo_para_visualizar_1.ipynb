```python
import pandas as pd
```

```python
db= pd.read_csv('all-weeks-global.csv')
```

```python
weeks= db['week'].unique()
```

```python
db=db[db["weekly_rank"].isin([1])]
```

```python
db_series= db[db['category']=="TV (English)"]
```

```python
db_series
```

```python
filtered_series = db_series[db_series['season_title'].str.contains(r'Season [2-9]|Season [1-9]\d', regex=True)]
```

```python
filtered_series
```

```python


week_sequel=[]

for week in weeks:
  series_week = filtered_series[filtered_series['week'] == week]
  if not series_week.empty:
    week_sequel.append(week)
```

```python
len(week_sequel)
```

```python
len(weeks)
```

```python
top=pd.DataFrame()
```

```python
top.index=weeks
```

```python
top
```

```python
top["sequel"]=0
```

```python
top
```

```python
top["week"]=weeks
```

```python
top
```

```python
top=top.reset_index(drop=True)
```

```python
top
```

```python
top["sequel"]=top["week"].apply(lambda x: 1 if x in week_sequel else 0)
```

```python
top
```

```python
db_series=db_series.reset_index(drop=True)
top["season_title"]=db_series["season_title"]
top["weekly_hours_viewed"]=db_series["weekly_hours_viewed"]
```

```python
top
```

```python
top=top.iloc[::-1].reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt
```

```python

week = top.index
hours = top["weekly_hours_viewed"]

plt.figure(figsize=(10, 6))

plt.plot(week, hours, marker='o', linestyle='-', color='b', label='Horas por semana')

plt.xlabel('Semana')
plt.ylabel('Horas')
plt.title('Horas por semana')
plt.grid(True)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt


week = top.index
hours = top['weekly_hours_viewed']
sequel = top['sequel']

plt.figure(figsize=(10, 6))


plt.scatter(week[sequel == 1], hours[sequel == 1], color='red', label='Secuela')
plt.scatter(week[sequel == 0], hours[sequel == 0], color='blue', label='No secuela')

plt.plot(week, hours, color='black', alpha=0.3)

plt.ticklabel_format(style='plain', axis='y')


plt.xlabel('Semana')
plt.ylabel('Horas')
plt.title('Horas vistas del Top 1 Semanal de series de habla inglesa en Netflix')

plt.legend()

# Mostrar el grfico
# plt.grid(True)
# plt.show()

html_fig = mpld3.fig_to_html(plt.gcf())

with open('grafico_interactivo.html', 'w') as f:
    f.write(html_fig)
```
